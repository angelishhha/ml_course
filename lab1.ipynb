```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
sns.set(style="ticks")

data = pd.read_csv('artists.csv', sep=",")

data.head()
data.shape
total_count = data.shape[0]
print('Всего строк: {}'.format(total_count))

data.columns
data.dtypes
for col in data.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
    
data.describe()

data['paintings'].unique()

id = data['id']
paintings = data['paintings']
plt.scatter(id, paintings, edgecolors='r')
plt.xlabel('ID')
plt.ylabel('PAINTINGs')
plt.title('***')
plt.show()
#fig, ax = plt.subplots(figsize=(10,10)) 
#plt.scatterplot(ax=ax, x='id', y='paintings', data=data)

fig, ax = plt.subplots(figsize=(10,10)) 
sns.distplot(data['paintings'])

sns.jointplot(x='id', y='paintings', data=data)
sns.jointplot(x='id', y='paintings', data=data, kind="hex")
sns.jointplot(x='id', y='paintings', data=data, kind="kde")

sns.pairplot(data)


sns.pairplot(data)
```
